```python
import numpy as np


class LSTM:

    def __init__(self):
        self.Wf, self.Uf, self.bf = 0.5, 0.1, 0
        self.Wi, self.Ui, self.bi = 0.6, 0.2, 0
        self.Wc, self.Uc, self.bc = 0.7, 0.3, 0
        self.Wo, self.Uo, self.bo = 0.8, 0.4, 0
        self.Wy, self.by = 4, 0
        self.h = 0
        self.C = 0

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def tanh(self, x):
        return np.tanh(x)

    def step(self, x, t):
        f = self.sigmoid(self.Wf * x + self.Uf * self.h + self.bf)
        i = self.sigmoid(self.Wi * x + self.Ui * self.h + self.bi)
        c_bar = self.tanh(self.Wc * x + self.Uc * self.h + self.bc)
        self.C = f * self.C + i * c_bar
        o = self.sigmoid(self.Wo * x + self.Uo * self.h + self.bo)
        self.h = o * self.tanh(self.C)

        print("Time Step:", t)
        print("Input:", x)
        print("Forget Gate:", round(f,4))
        print("Input Gate:", round(i,4))
        print("Candidate State:", round(c_bar,4))
        print("Cell State:", round(self.C,4))
        print("Output Gate:", round(o,4))
        print("Hidden State:", round(self.h,4))
        print()

    def run(self, sequence):
        for t, x in enumerate(sequence, 1):
            self.step(x, t)

    def predict(self):
        y_hat = self.Wy * self.h + self.by
        return y_hat


sequence = [1, 2, 3, 4]

model = LSTM()

model.run(sequence)

prediction = model.predict()

print("Final Prediction:", round(prediction,4))
```
